## CASO A ##

1. Función de Transferencia:
La expresión de V0 resulta:
   
    $V0 = \frac{ZrVi}{ Zr + Zc + Zl}$
   
Sustituyendo las impedancias en terminos de S:

  $V0 = \frac{RVi}{ (R + 1/SC + SL }$

 $ H(s) = V0 (s) / Vi (s) $
 
 $ H(s) = \frac{ R }{ (R + 1/SC + SL })$
 
 Por lo que H(s) resulta:
 
  $\frac{SRC}{RSC + 1 + S^2CL}$

2. Calcular respuesta en magnitud y fase.

Sustituimos s=σ+jw con σ=0 (Analizamos solo el eje imaginario)

$ |T(s)|(jw) |= | \frac{RCwj}{RCwj + 1 - w^2CL}  |$

$| T(jw) | = \frac{RCw}{\sqrt{(RCw)^2+(1-LCw^2)^2}}$

$ R/L = w_0/Q ; 1/LC = w_0^2  ->  RC = R/L CL = w_0/Q CL ; CL = 1/w_0^2 L $

$| T(jw) | = \frac{w_0/Q LCw}{\sqrt{(w_0/Q LCw)^2+(1-w_0^2 LCw^2)^2}}$

$| T(jw) | = \frac{w/w_0 Q}{\sqrt{(w/w_0)^2 Q^2+(1-(w/w_0)^2)^2}}$

Para obtener la fase, separamos parte real de imaginaria de T(jw):

$ T(jw) = \frac{RCwj}{RCwj + 1 - w^2CL} * \frac{RCwj - (1 - w^2CL)}{RCwj - (1 - w^2CL)} $

$ T(jw) = \frac{RCwj-RLC^2w^3j+(RCw)^2}{(RCw)^2 + (1 - w^2CL)^2} $

$ T(jw) = \frac{(RCw)^2}{(RCw)^2 + (1 - w^2CL)^2} + \frac{RCw-RLC^2w^3}{(RCw)^2 + (1 - w^2CL)^2}j $

$ T(jw) = \frac{(w/w_0)^2 Q^2}{(w/w_0)^2 Q^2 + (1 - (w/w_0)^2)^2} + j \frac{(w/w_0) Q (1 - (w/w_0)^2)}{(w/w_0)^2 Q^2 + (1 - (w/w_0)^2)^2} $

Consideramos:

$ θ(w) = \tan^{-1} (\frac{Im(T(jw))}{Re(T(jw))}) $

$ θ(w) = \tan^{-1} ( \frac{\frac{(w/w_0) Q (1 - (w/w_0)^2)}{(w/w_0)^2 Q^2 + (1 - (w/w_0)^2)^2}}{\frac{(w/w_0)^2 Q^2}{(w/w_0)^2 Q^2 + (1 - (w/w_0)^2)^2}} $

$ θ(w) = \tan^{-1} ( \frac{(w/w_0) Q (1 - (w/w_0)^2)}{(w/w_0)^2 Q^2} ) $

$ θ(w) = \tan^{-1} ( \frac{1 - (w/w_0)^2}{(w/w_0) Q} )$


3. Tipo de filtro:
   Al analizar los bordes (s-> ∞ y s->0) observamos lo siguiente:
   
 $\lim_{w \to \infty} |T(jw)| = \lim_{w \to \infty} | \frac{w/w_0 Q}{\sqrt{(w/w_0)^2 Q^2+(1-(w/w_0)^2)^2}}| = 0 $
 
 $\lim_{s \to 0} |T(jw)| = \lim_{w \to 0} | \frac{w/w_0 Q}{\sqrt{(w/w_0)^2 Q^2+(1-(w/w_0)^2)^2}}| = 0 $

La respuesta en magnitud tiende a 0 en ambos bordes, lo que significa que el filtro atenúa para frecuencias altas y bajas, por lo que es un filtro pasa banda.
 




## CASO B ##

1. Función de Transferencia:
La expresión de V0 resulta:
   
    $V0 = \frac{Z equivalente paralelo}{ Z total}$
   
Sustituyendo las impedancias en terminos de S:

  $V0 = \frac{\frac{RLVi}{R+SL}}{\frac{RL}{R+SL} + 1/SC }$

 $ H(s) = V0 (s) / Vi (s) $
 
 $H(s) = \frac{\frac{RL}{R+SL}}{\frac{RL}{R+SL} + 1/SC }$
 
 Por lo que H(s) resulta:
 
  $H(s) =\frac{RLSC}{RLSC + R+ SL}$

2. Calcular respuesta en magnitud y fase.

Sustituimos s=σ+jw con σ=0 (Analizamos solo el eje imaginario)

$|T(jw)| = | \frac{jwCRL}{RLCjw + R + jwL}| $

$|T(jw)| = \frac{wCRL}{\sqrt{(w(CRL + L))^2 + R^2}} $

$|T(jw)| = \frac{1}{Q w_0} \frac{w}{\sqrt{( w(1 + \frac{1}{Q w_0}))^2 + (\frac{w_0}{Q})^2}}$

Respuesta en fase:

$ θ(w) = \tan^{-1} (\frac{Im(T(jw))}{Re(T(jw))}) $

Separamos parte real de imaginaria en T(jw) 

$T(jw) = \frac{jwCRL}{jw (CRL + L) + R}  \frac{-jw (CRL + L) + R}{-jw (CRL + L) + R} $

$T(jw) = \frac{-w^2CRL (CRL + L) j + wCRLR}{w^2 (CRL + L)^2 + R^2} $

$T(jw) = \frac{wCRLR}{w^2 (CRL + L)^2 + R^2} + j \frac{-w^2 CRL (CRL + L)}{w^2 (CRL + L)^2 + R^2} $

$ T(jw) = \frac{w \frac{1}{Q w_0} \frac{w_0}{Q}}{w^2 (1 + \frac{1}{Q w_0})^2 + (\frac{w_0}{Q})^2} + j \frac{-w^2 \frac{1}{Q w_0} (1 + \frac{1}{Q w_0})}{w^2(1 + \frac{1}{Q w_0})^2 + (\frac{w_0}{Q})^2} $

$ θ(w) = \tan^{-1} (\frac{\frac{-w^2 \frac{1}{Q w_0} (1 + \frac{1}{Q w_0})}{w^2(1 + \frac{1}{Q w_0})^2 + (\frac{w_0}{Q})^2}}{\frac{w \frac{1}{Q w_0} \frac{w_0}{Q}}{w^2 (1 + \frac{1}{Q w_0})^2 + (\frac{w_0}{Q})^2}}) = \tan^{-1} (\frac{-w^2 \frac{1}{Q w_0} (1 + \frac{1}{Q w_0})}{w \frac{1}{Q w_0} \frac{w_0}{Q}}) $

$ θ(w) = \tan^{-1} (\frac{-wQ(1 + \frac{1}{Q w_0})}{w_0}) $ 

3. Tipo de filtro:
   Al analizar los bordes (s-> ∞ y s->0) observamos lo siguiente:
    
$ \lim_{w \to \infty} |T(jw)| = \lim_{w \to \infty} \frac{\frac{1}{Q w_0} w}{\sqrt{( w(1 + \frac{1}{Q w_0}))^2 +(\frac{w_0}{Q})^2}} = 1 $

$ \lim_{w \to 0} | T(jw)| = \lim_{w \to 0} \frac{\frac{1}{Q w_0} w}{\sqrt{( w(1 + \frac{1}{Q w_0})^2 + (\frac{w_0}{Q})^2}} = 0 $

La magnitud de la ganancia tiende a 0 para w tendiendo a 0, esto sugiere que las frecuencias muy bajas son atenuadas. Por otra parte, cuando la frecuencia tiende a valores altos, la ganancia tiende a 1, es dcir que las señales de alta frecuencia no son atenuadas. Por lo tanto, se concluye que la imagen b es un filtro pasa altos.

```python
# -*- coding: utf-8 -*-
"""
Created on Sun Mar 16 23:03:53 2025

@author: iremo
"""

import numpy as np
import matplotlib.pyplot as plt

Q = 1
w0= 1
# se analiza un intervalo dos décadas antes y después de la frecuencia de corte. (w/100 hasta w0*100)
w= np.logspace(np.log10(w0/100), np.log10(100*w0), 10000)

def funciona(Q, w0):
    w = np.logspace(np.log10(w0/100), np.log10(100*w0), 10000)
    Ta = (w / (w0 * Q)) / np.sqrt((w / w0)**2 * Q**2 + (1 - (w / w0)**2)**2)
    return w, Ta
def fasea(Q,w0):
    w = np.logspace(np.log10(w0/100), np.log10(100*w0), 10000)
    fa = np.arctan((1 - (w / w0)**2) / ((w / w0) * Q))
    return w, fa

w, Ta =funciona(Q, w0)
w, fa =fasea(Q,w0)

plt.figure(1)
plt.semilogx(w, 20*np.log10(Ta), label="Magnitud")
plt.xlabel("Frecuencia (rad/s)")
plt.ylabel("Ganancia (dB)")
plt.title("Diagrama de Bode - Magnitud A")
plt.grid()

plt.figure(2)
plt.semilogx(w, fa)  
plt.ylabel("Fase (radianes)")
plt.xlabel("Frecuencia (rad/s)")
plt.title("Diagrama de Bode - Fase A")
plt.grid()

plt.show()
```

```python

import numpy as np
import matplotlib.pyplot as plt

Q = 1
w0= 1
# se analiza un intervalo dos décadas antes y después de la frecuencia de corte. (w/100 hasta w0*100)
w= np.logspace(np.log10(w0/100), np.log10(100*w0), 10000)

def funcionb(Q, w0):
    w = np.logspace(np.log10(w0/100), np.log10(100*w0), 10000)
    Tb = (1 / (Q * w0)) * (w / np.sqrt((w * (1 + 1 / (Q * w0)))**2 + (w0 / Q)**2))
    return w, Tb
    
def faseb(Q,w0):
    w = np.logspace(np.log10(w0/100), np.log10(100*w0), 10000)
    fb = np.arctan(-w * Q * (1 + 1 / (Q * w0)) / w0)
    return w, fb
    
w, Tb =funcionb(Q, w0)
w, fb =faseb(Q,w0)

plt.figure(3)
plt.semilogx(w, 20*np.log10(Tb), label="Magnitud")
plt.xlabel("Frecuencia (rad/s)")
plt.ylabel("Ganancia (dB)")
plt.title("Diagrama de Bode - Magnitud B")
plt.grid()

plt.figure(4)
plt.semilogx(w, fb)  
plt.ylabel("Fase (radianes)")
plt.xlabel("Frecuencia (rad/s)")
plt.title("Diagrama de Bode - Fase B")
plt.grid()

plt.show()
    
```

La imagen A representa un filtro pasa banda.
