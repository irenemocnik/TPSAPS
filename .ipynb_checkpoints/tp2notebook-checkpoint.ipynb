1. Funci√≥n de Transferencia:

 $ H(s) = V0 (s) / Vi (s) $
 
 $ H(s) = \frac{ R }{ (R + 1/SC + SL })$
 
 Por lo que H(s) resulta:
 
  $\frac{SRC}{RSC + 1 + S^2CL}$

2. Calcular respuesta en magnitud y fase.

Sustituimos s=œÉ+jw con œÉ=0 (Analizamos solo el eje imaginario)

$ |T(s)|(jw) |= | \frac{RCwj}{RCwj + 1 - w^2CL}  |$

$| T(jw) | = \frac{RCw}{\sqrt{(RCw)^2+(1-LCw^2)^2}}$

$ T(jw) = \frac{RCwj}{RCwj + 1 - w^2CL} * \frac{RCwj - (1 - w^2CL)}{RCwj - (1 - w^2CL)} $

$ T(jw) = \frac{RCwj-RLC^2w^3j+(RCw)^2}{(RCw)^2 + (1 - w^2CL)^2} $

$ Œ∏(w) = \tan^{-1} ( \frac{RCw-RLC^2w^3}{(RCw)^2} ) $

## CASO B ##

1. Funci√≥n de Transferencia:
La expresi√≥n de V0 resulta:
   
    $V0 = \frac{Z equivalente paralelo}{ Z total}$
   
Sustituyendo las impedancias en terminos de S:

  $V0 = \frac{\frac{RLVi}{R+SL}}{\frac{RL}{R+SL} + 1/SC }$

 $ H(s) = V0 (s) / Vi (s) $
 
 $H(s) = \frac{\frac{RL}{R+SL}}{\frac{RL}{R+SL} + 1/SC }$
 
 Por lo que H(s) resulta:
 
  $H(s) =\frac{RLSC}{RLSC + R+ SL}$

2. Calcular respuesta en magnitud y fase.

Sustituimos s=œÉ+jw con œÉ=0 (Analizamos solo el eje imaginario)

$|T(jw)| = | \frac{jwCRL}{RLCjw + R + jwL}| $

$|T(jw)| = \frac{wCRL}{\sqrt{(w(CRL + L))^2 + R^2}} $

Respuesta en fase:

$ Œ∏(w) = \tan^{-1} (\frac{Im(T(jw))}{Re(T(jw))}) $

Separamos parte real de imaginaria en T(jw) 

$T(jw) = \frac{jwCRL}{jw (CRL + L) + R}  \frac{-jw (CRL + L) + R}{-jw (CRL + L) + R} $

$T(jw) = \frac{-w^2CRL (CRL + L) j + wCRLR}{w^2 (CRL + L)^2 + R^2} $

$T(jw) = \frac{wCRLR}{w^2 (CRL + L)^2 + R^2} + j \frac{-w^2 CRL (CRL + L)}{w^2 (CRL + L)^2 + R^2} $

3. Tipo de filtro:
   Al analizar los bordes (s-> ‚àû y s->0) observamos lo siguiente:
   
 $\lim_{s \to \infty} |T(s)| = \lim_{s \to \infty} | \frac{sCRL}{\sqrt{(s(CRL + L))^2 + R^2}} | = \frac{CRL}{CRL + L} $
 
 $\lim_{s \to 0} |T(s)| = \lim_{s \to 0} | \frac{sCRL}{\sqrt{(s(CRL + L)^2 + R^2}}| = 0 $



```python
import numpy as np
import matplotlib.pyplot as plt

r=5.00
c=0.000002
l=0.5
wf=300.00

def funciona(r,c,l):
    w = np.arange(start = 0, stop = wf, step = 20)
    Ta = w*c*r/(np.sqrt(w*w*r*r*c*c+(1-l*c*w*w)*(1-l*c*w*w)))
    return w, Ta
w, Ta =funciona(r, c, l)
plt.figure(1)
plt.plot(w, 20*np.log10(Ta), label="Magnitud")

def funcionb(r,c,l):
    w = np.arange(start = 0, stop = wf, step = 20)
    Tb = w*c*r*l/(np.sqrt(w*w*(c*r*l + l)*(c*r*l + l) + r*r))
    return w, Tb
w, Tb =funcionb(r, c, l)
plt.figure(2)
plt.plot(w, 20*np.log10(Tb), label="Magnitud")
```
